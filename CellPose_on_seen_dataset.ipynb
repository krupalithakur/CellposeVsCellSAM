```python
import requests, json, os

# Path to the JSON file
json_url = "https://livecell-dataset.s3.eu-central-1.amazonaws.com/LIVECell_dataset_2021/annotations/LIVECell_single_cells/a172/train.json"
data = requests.get(json_url).json()

os.makedirs("LIVECell_A172_train", exist_ok=True)

for img in data["images"]:
    url = img["url"]
    filename = img["file_name"]
    filepath = f"LIVECell_A172_train/{filename}"

    img_data = requests.get(url).content
    with open(filepath, "wb") as f:
        f.write(img_data)

print("Downloaded", len(data["images"]), "images ✅")
```

```python
import json
import os
import cv2
import numpy as np
from pycocotools.coco import COCO
from tqdm import tqdm

# Paths
json_path = "/content/LIVECell_dataset_2021/annotations/LIVECell/livecell_coco_train.json"
img_folder = "/content/LIVECell_dataset_2021/images/images/livecell_train_val_images"
mask_output_folder = "/content/LIVECell_dataset_2021/masks/train"

os.makedirs(mask_output_folder, exist_ok=True)

# Load COCO annotations
coco = COCO(json_path)
img_ids = coco.getImgIds()

for img_id in tqdm(img_ids):
    img_info = coco.loadImgs(img_id)[0]
    img_name = img_info["file_name"]
    width, height = img_info["width"], img_info["height"]

    # Create empty mask
    mask = np.zeros((height, width), dtype=np.uint8)

    ann_ids = coco.getAnnIds(imgIds=img_id)
    anns = coco.loadAnns(ann_ids)

    for ann in anns:
        segmentation = coco.annToMask(ann)
        mask = np.maximum(mask, segmentation * 255)  # merge masks

    # Save mask
    mask_name = os.path.splitext(img_name)[0] + "_masks.png"
    cv2.imwrite(os.path.join(mask_output_folder, mask_name), mask)

print("✅ All masks saved in:", mask_output_folder)
```
